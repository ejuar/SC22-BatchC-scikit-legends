```python
music_df.dropna(inplace = True)
```

```python
music_df.info()
```

```python
# removing missing values
# remove ? from tempo
music_df = music_df[music_df['tempo'] != '?']
```

```python
music_df.info()
```

```python
music_df[music_df['duration_ms'] != -1]
```

```python
music_df['artist_name'].value_counts()
```

```python
columns_to_drop = ['instance_id','track_name']
music_df.drop(columns_to_drop, axis-1, inplace=True)
```

```python
fig = px.histogram(music_df, x='popularity')
fig.show()
```

```python
fig = px.scatter(music_df, x = 'tempo')
fig.show()
```

```python
fig = px.violin(music_df, x = 'loudness')
fig.show()
```

```python
fig = px.funnel(music_df, x = 'popularity')
fig.show()
```

```python
fig = px.pie(music_df, x = 'popularity')
fig.show()
```
