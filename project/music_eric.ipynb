```python
import pandas as pd
import plotly.express as px
```

```python
data_path = 'data/music.csv'
music_df = pd.read_csv(data_path)
```

```python
music_df.dropna(inplace=True)
```

```python
music_df.info()
```

```python
# removing missing values
# remove ? from tempo
music_df = music_df[music_df['tempo'] != '?']
```

```python
music_df.info()
```

```python
# removing missing values
# remove -1 from duration_df
music_df = music_df[music_df['duration_ms'] != -1.]
```

```python
music_df.info()
```

```python
music_df['music_genre'].value_counts()
```

```python
music_df['artist_name'].value_counts()
```

```python
columns_to_drop = ['instance_id','track_name']
music_df.drop(columns_to_drop, axis=1, inplace=True)
```

```python
music_df.info()
```

```python
fig = px.histogram(music_df, x='popularity')
fig.show()
```
