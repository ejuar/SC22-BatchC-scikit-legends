```python
music_df.dropna(inplace=True)
```

```python
music_df.info()
```

```python
music_df = music_df[music_df['tempo'] != '?']
```

```python
music_df.info()
```

```python
music_df = music_df[music_df['duration_ms'] != -1]
```

```python
music_df.info()
```

```python
music_df['music_genre'].value_counts()
```

```python
music_df['artist_name'].value_counts()
```

```python
columns_to_drop = ['instance_id','track_name']
music_df.drop(columns_to_drop, axis=1, inplace=True)
```

```python
music_df.info()
```

```python
fig = px.box(music_df, x='popularity', y='music_genre')
fig.show()
```

```python
fig = px.histogram(music_df, x='music_genre')
```

```python
fig = px.pie(music_df, values='popularity', names='music_genre')
fig.show()
```

```python
fig = px.violin(music_df, y="obtained_date")
fig.show()
```

```python
fig = px.scatter(music_df, x="energy", y="danceability")
fig.show()
```

```python
fig = px.line(music_df, x="liveness", y="loudness")
fig.show()
```
